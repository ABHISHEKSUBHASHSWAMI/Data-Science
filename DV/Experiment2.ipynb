## **1. Bar Chart**

### Matplotlib 

```python
import matplotlib.pyplot as plt

categories = ['Category A', 'Category B', 'Category C']
values = [10, 15, 8]

plt.barh(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart - Matplotlib')
plt.show()
```

### Seaborn

```python
import seaborn as sns

categories = ['Category A', 'Category B', 'Category C']
values = [10, 15, 8]

sns.barplot(x=values, y=categories)
plt.xlabel('Values')
plt.ylabel('Categories')
plt.title('Horizontal Bar Chart - Seaborn')
plt.show()
```

## **3. Lollipop Chart**

### Matplotlib

```python
import matplotlib.pyplot as plt

categories = ['Category A', 'Category B', 'Category C']
values = [10, 15, 8]

plt.stem(categories, values, basefmt='b-', markerfmt='s')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Lollipop Chart - Matplotlib')
plt.show()
```

### Seaborn 

```python
import seaborn as sns

categories = ['Category A', 'Category B', 'Category C']
values = [10, 15, 8]

sns.pointplot(x=categories, y=values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Lollipop Chart - Seaborn')
plt.show()
```

## **4. Bullet Chart**

### Plotly

```python
import plotly.graph_objects as go

fig = go.Figure(go.Indicator(
    mode = "number+gauge+delta", value = 220,
    domain = {'x': [0.1, 1], 'y': [0, 1]},
    title = {'text' :"<b>Example 1</b>"},
    delta = {'reference': 200},
    gauge = {
        'shape': "bullet",
        'axis': {'range': [None, 300]},
        'threshold': {
            'line': {'color': "red", 'width': 2},
            'thickness': 0.75,
            'value': 280},
        'steps': [
            {'range': [0, 150], 'color': "lightgray"},
            {'range': [150, 250], 'color': "gray"}]}))
fig.update_layout(height = 250)
fig.show()
```

### Matplotlib and Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import FuncFormatter

limits = [80, 100, 150]
data_to_plot = ("Example 2", 105, 120)

palette = sns.color_palette("Blues_r", len(limits))

fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.set_yticks([1])
ax.set_yticklabels([data_to_plot[0]])

prev_limit = 0
for idx, lim in enumerate(limits):
    ax.barh([1], lim-prev_limit, left=prev_limit, height=15, color=palette[idx])
    prev_limit = lim

# Draw the value we're measuring
ax.barh([1], data_to_plot[1], color='black', height=5)

ax.axvline(data_to_plot[2], color="gray", ymin=0.10, ymax=0.9)
```

## **5. Dubmbell Chart**

### Matplotlib

```python
import matplotlib.pyplot as plt

categories = ['Category A', 'Category B', 'Category C']
values_start = [10, 15, 35]
values_end = [15, 20, 25]

plt.hlines(y=categories, xmin=values_start, xmax=values_end, color='blue')
plt.scatter(values_start, categories, color='blue', marker='o')
plt.scatter(values_end, categories, color='blue', marker='o')
plt.xlabel('Values')
plt.title('Dumbbell Chart - Matplotlib')
plt.show()
```
