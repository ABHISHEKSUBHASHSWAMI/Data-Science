````python
#Lollipop
import matplotlib.pyplot as plt

categories = ['Category A', 'Category B', 'Category C']
values = [10, 15, 8]

plt.stem(categories, values, basefmt='b-', markerfmt='o')
plt.setp(plt.gca().lines, markersize=20)  # Increasing the marker size to 20

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Lollipop Chart - Matplotlib')
plt.show()
plt.close()
````

3. **Creating the stem plot:**

    - `plt.stem()` creates a stem plot (lollipop chart) with the given categories and values.

    - `delta` specifies the reference value for the delta calculation. The difference between the current value (220) and the reference value (200) will be displayed as the delta.

7. **Configuring the gauge (bullet chart):**

    gauge = {

### Usage of Bullet Chart and its Components:

A bullet chart is a variation of a bar chart developed by Stephen Few. It is useful for displaying performance data compared to a target or benchmark value. Bullet charts are particularly effective for dashboard visualizations where space is limited.

**Components of a Bullet Chart:**
- **Measure (Value):** The actual value of the metric being tracked, displayed as a horizontal bar.
- **Comparative Measure (Target/Threshold):** A line indicating the target or threshold value for the metric.
- **Ranges (Steps):** Background color bands that provide qualitative ranges for the value (e.g., poor, satisfactory, good).
- **Delta:** The difference between the actual value and the target or reference value, often shown as an additional number or indicator.

In this example:
- The current value (220) is shown as a horizontal bar.
- The threshold (280) is shown as a red vertical line.
- The background ranges from 0 to 150 (light gray) and from 150 to 250 (gray) provide context for evaluating the value.
- The delta shows the difference between the current value (220) and the reference value (200).

````python
#Contour
import numpy as np
import matplotlib.pyplot as plt

# Create some sample data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create a contour plot
plt.figure(figsize=(8, 6))
contour = plt.contour(X, Y, Z, cmap='viridis')

# Add labels and a colorbar
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Contour Plot Example')
plt.colorbar(contour, label='Z-axis')

# Show the plot
plt.show()
plt.close()
````

    ```python
    import numpy as np
    import matplotlib.pyplot as plt
    ```

    - `matplotlib.pyplot` is imported as `plt` for plotting and visualizations.

2. **Creating the sample data:**
    ```python

    y = np.linspace(-5, 5, 100)
    X, Y = np.meshgrid(x, y)

    ```
    - `x = np.linspace(-5, 5, 100)` creates an array of 100 evenly spaced values from -5 to 5 for the x-axis.
    - `y = np.linspace(-5, 5, 100)` creates an array of 100 evenly spaced values from -5 to 5 for the y-axis.
    - `X, Y = np.meshgrid(x, y)` creates a coordinate grid from the `x` and `y` arrays. `X` and `Y` are 2D arrays representing the grid coordinates.
    - `Z = np.sin(np.sqrt(X**2 + Y**2))` calculates the Z values for each (X, Y) pair using the function `Z = sin(sqrt(X^2 + Y^2))`. This creates a wave-like pattern.

3. **Creating the contour plot:**
    ```python

    - `plt.figure(figsize=(8, 6))` creates a new figure with a specified size of 8x6 inches.
    - `contour = plt.contour(X, Y, Z, cmap='viridis')` creates a contour plot with the data arrays `X`, `Y`, and `Z`. The `cmap='viridis'` parameter sets the colormap to 'viridis', which is a color gradient used to represent different levels in the contour plot.

4. **Adding labels and a colorbar:**
    ```python

# 5. Implementation of part to whole and hierarchical based graphs and charts : Stacked Bar & Column, Pie

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Data
groups = ['G1', 'G2', 'G3', 'G4', 'G5']
values1 = [12, 19, 14, 27, 16]
values2 = [21, 30, 15, 17, 20]

# Creating a DataFrame
df = pd.DataFrame({
    'Group': groups,
    'Values1': values1,
    'Values2': values2
})

# Creating the bar plot
fig, ax = plt.subplots()

# Plotting the first set of bars
bars1 = sns.barplot(x=values1, y=groups, color='green', label='Values1', orient='h')

# Adding the second set of bars
bars2 = sns.barplot(x=values2, y=groups, left=values1, color='purple', label='Values2', orient='h')

# Adding labels and title
plt.xlabel('Values')
plt.ylabel('Groups')
plt.title('Stacked Horizontal Bar Chart')

# Adding a legend
plt.legend()

# Displaying the plot
plt.show()
plt.close()
````

# 12. Silhouette Analysis - Selecting the number of clusters Visually

````python
#Silhouette Analysis
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from yellowbrick.cluster import silhouette_visualizer
from warnings import filterwarnings

filterwarnings("ignore")


# Create sample data with 4 clusters
X, y = make_blobs(n_samples=500, centers=4, cluster_std=1.0, random_state=42)

# Range of number of clusters to test
cluster_range = range(2, 7)

# List to store the average silhouette scores
silhouette_scores = []

# Perform KMeans clustering and compute silhouette scores
for n_clusters in cluster_range:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    
    cluster_labels = kmeans.fit_predict(X)
    silhouette_avg = silhouette_score(X, cluster_labels)
    silhouette_scores.append(silhouette_avg)
    silhouette_visualizer(kmeans, X, colors='yellowbrick')
    print(f"For n_clusters = {n_clusters}, average silhouette score is {silhouette_avg:.4f}")

# Determine the optimal number of clusters
optimal_n_clusters = cluster_range[np.argmax(silhouette_scores)]
print(f"The optimal number of clusters (k) is {optimal_n_clusters}")
````

This code performs silhouette analysis to determine the optimal number of clusters for a dataset:

1. **Importing Libraries**:
    - `numpy`: Numerical computing library in Python.

    - `KMeans`: Clustering algorithm from scikit-learn.

# 14. Time Series Analysis & Forecasting

````python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Define the stock ticker symbol and time period
stock_symbol = 'NHPC.NS'
start_date = '2020-01-01'
end_date = '2024-05-30'

# Download historical stock data
stock_data = yf.download(stock_symbol, start=start_date, end=end_date)

# Extract the closing prices
closing_prices = stock_data['Close']

# Create a date range with the same length as the closing prices
date_range = pd.date_range(start=start_date, end=end_date, freq='B')

# Adjust the frequency of the date range to match the frequency of the stock data
date_range = date_range[:len(closing_prices)]

# Set the index of the closing prices
closing_prices.index = date_range

# Calculate Simple Moving Average (SMA)
sma = closing_prices.rolling(window=20).mean()

# Calculate Exponential Moving Average (EMA)
ema = closing_prices.ewm(span=20, adjust=False).mean()

# Calculate Cumulative Moving Average (CMA)
cma = closing_prices.expanding(min_periods=20).mean()

# Plotting actual, SMA, EMA, and CMA
plt.figure(figsize=(10, 6))
plt.plot(closing_prices.index, closing_prices, color='blue', label='Actual Prices')
plt.plot(sma.index, sma, color='green', label='SMA (20)')
plt.plot(ema.index, ema, color='orange', label='EMA (20)')
plt.plot(cma.index, cma, color='purple', label='CMA')
plt.title('Actual Prices with Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price (INR)')
plt.legend()
plt.xticks(rotation=90)  # Rotate x-axis labels by 45 degrees
plt.grid(True)
plt.show()
````

````python
#forecasting
import numpy as np
from keras.models import Sequential
from keras.layers import Dropout, Dense, GRU
from keras.callbacks import ModelCheckpoint, EarlyStopping
import keras
import matplotlib.pyplot as plt

def format(series: np.array = np.array([]), look_back: int = 5):
    """
    Create a (X, Y) feature-output dataset for ML workflow
    :param series: Numpy (1, ) array
    :param look_back: The number of previous features to be considered for forecasting
    :return: X and Y modelling parameters
    """
    X = []
    Y = []
    for n in range(len(series)):
        k = n + look_back
        if k >= len(series):
            break
        Xtn = series[n:k]
        if len(Xtn) != look_back:
            break
        Ytn = series[k]
        X.append(Xtn)
        Y.append(Ytn)
    return np.array(X), np.array(Y)

def gru_prediction(stock_data):
    model_filepath='gru_model.keras'

    # Calculate the mean of the 'Close' series
    mean = np.mean(stock_data['Close'].values)

    # Adjust the series by downscaling it using the mean
    adjusted_series = stock_data['Close'].values - mean

        # Define the look-back period you want to use
    look_back = 30  # This can be adjusted according to your needs

    # Format the adjusted series using the 'format' function
    X, y = format(adjusted_series, look_back=look_back)

    # Reshape the feature set (X) to be suitable for LSTM models
    Xr = X.reshape((X.shape[0], 1, X.shape[1]))

    # Create the Sequential model with a name
    model = Sequential(name="Proposed_GRU")
    model.add(GRU(units=512, activation='relu', input_shape=(1,look_back), return_sequences=True, name="input"))
    model.add(Dropout(0.2))
    model.add(GRU(units=512, activation='relu', name="gru"))
    model.add(Dropout(0.2))
    model.add(Dense(1, name="output"))
    model.compile(loss='mse', optimizer='adam')

    # Save the best model based on validation loss
    checkpoint_callback = ModelCheckpoint(
        filepath=model_filepath,
        save_best_only=True,
        monitor='val_loss',  # Metric to monitor
        mode='min',  # Minimize the monitored metric
        verbose=1  # Verbosity level
    )

    # Create an EarlyStopping callback to stop training when validation loss does not improve
    early_stopping = EarlyStopping(
        monitor='val_loss',  # Metric to monitor
        mode='min',  # Minimize the monitored metric
        patience=10,  # Number of epochs to wait before stopping if no progress
        verbose=1  # Verbosity level
    )

    # Add the callbacks to a list
    callbacks = [checkpoint_callback, early_stopping]

    # Train the model
    model.fit(Xr, y, epochs=200,  batch_size=128,  validation_split=0.2, callbacks=callbacks)

    #load model
    model=keras.models.load_model(model_filepath)

    days=30
    mean_added_pred = []
    x = Xr[-1]  # Initially take last training input for predicting its immediate next value

    for i in range(days):
      x = x.reshape((1, 1, look_back))  # Prepare the input for a tensorflow model using reshaping
      prediction = model.predict(x, verbose=False)   # Predict the immediate next value
      x = x.ravel()  # Ravel method converts the 3D array to single dimension array 
      x = np.delete(x, 0)  # Now, we delete first item of the original input
      x = np.append(x, prediction[0][0])  # and append the predicted value to use it for next predictions
      mean_added_pred.append(prediction[0][0])   # append the predicted value to 'predictions' list

    mean_added_pred+=mean
    y+=mean
    fig, ax = plt.subplots(figsize=(16, 8))

# Plot the forecasted series and the original series
    ax.plot(np.append(y[-100:], mean_added_pred), color='r', label="Forecasted series")
    ax.plot(y[-100:], color='b', label="Original series")

    # Set labels and legend
    ax.set_ylabel("closing price (Rs.)")
    ax.set_xlabel("Days")
    ax.legend()

    # Set the title
    ax.set_title("Next 30-days forecast using GRU")

    # Add a grid to the plot
    ax.grid(True, color='gray', linestyle='--', linewidth=0.5)
    plt.show()
    # Close the figure
    plt.close(fig)

gru_prediction(stock_data=stock_data)
````

# 15. Outlier Detection & Removal Analysis

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate some sample data
np.random.seed(0)
data = pd.Series(np.random.normal(loc=0, scale=1, size=1000))

# Define a function to detect outliers using Z-score
def detect_outliers_zscore(data, threshold=3):
    z_scores = (data - data.mean()) / data.std()
    outliers = data[abs(z_scores) > threshold]
    return outliers

# Detect outliers in the sample data
outliers = detect_outliers_zscore(data)

# Plot the distribution of the sample data
plt.figure(figsize=(10, 6))
plt.hist(data, bins=100, color='skyblue', edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Sample Data with Outliers')
plt.grid(True)

# Highlight the outliers in the plot
plt.scatter(outliers, np.zeros_like(outliers), color='red', label='Outliers')
plt.legend()
# Show the plot
plt.show()

# Remove outliers from the data
cleaned_data = data.drop(outliers.index)

# Plot the distribution of the cleaned data
plt.figure(figsize=(10, 6))
plt.hist(cleaned_data, bins=100, color='green', edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram of Cleaned Data (Outliers Removed using Z-Score)')
plt.grid(True)

# Show the plot
plt.show()
````
