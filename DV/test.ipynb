```python
x=np.random.rand(8)
y=np.random.rand(8)

sizes=np.random.rand(8)*1000

custom_colors=['red','blue','green','orange','brown','purple','pink','black']
custom_shapes=['o','s','D','^','v','H','p','*']
custom_edges=['-','--','-.',':','-','--','-.',':']

for i in range(len(x)):
    plt.scatter(x[i],y[i],s=sizes[i],c=custom_colors[i],marker=custom_shapes[i],edgecolors=custom_colors[-1*(i+1)],linestyle=custom_edges[i],linewidth=1)

plt.title('Customized BubblePlot')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create some sample data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create a contour plot
plt.figure(figsize=(8, 6))
contour = plt.contour(X, Y, Z, cmap='viridis')

# Add labels and a colorbar
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Contour Plot Example')
plt.colorbar(contour, label='Z-axis')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
groups = ['G1', 'G2', 'G3', 'G4', 'G5']
values1 = [12, 19, 14, 27, 16]
values2 = [21, 30, 15, 17, 20]

fig, ax = plt.subplots()

# Stacked bar chart
bar1 = ax.bar(groups, values1, label='Values 1')
bar2 = ax.bar(groups, values2, bottom=values1, label='Values 2')

# Add labels and title
ax.set_xlabel('Groups')
ax.set_ylabel('Values')
ax.set_title('Stacked Bar Chart Example')
ax.legend()

# Show the plot
plt.show()
```

```python
#Donut chart 

data=[40,30,20,10]
labels=['Apple','Banana','Guava','Grapes']
colours = ['red', 'blue', 'green','orange', 'white']
explode = (0.1, 0.1, 0.1, 0.1,0)
data.append(sum(data))
labels.append('')
plt.pie(data,labels=labels,explode=explode, startangle=360,wedgeprops=dict(width=0.3), colors=colours)
plt.title('Semi circle Donut Chart')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Generate some sample temporal data
date_rng = pd.date_range(start='2022-01-01', end='2022-01-10', freq='D')
values = np.random.randint(1, 100, size=len(date_rng))

# Create a DataFrame
df = pd.DataFrame({'Date': date_rng, 'Values': values})

# Plot the area chart
plt.figure(figsize=(10, 6))
plt.fill_between(df['Date'], df['Values'], color='blue', alpha=0.4, label='Temporal Data',) 

plt.title('Area Plot for Temporal Data')
plt.xlabel('Date')
plt.xticks(rotation=90)
plt.ylabel('Values')
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

date_rng = pd.date_range(start='2022-01-01', end='2022-01-10', freq='D')
values1 = np.random.randint(1, 50, size=len(date_rng))
values2 = np.random.randint(1, 50, size=len(date_rng))
colors = ['blue', 'orange']
df = pd.DataFrame({'Date': date_rng, 'Values1': values1, 'Values2': values2})
plt.figure(figsize=(10, 6))
plt.stackplot(df['Date'], df['Values1'], df['Values2'], labels=['Values1', 'Values2'], colors=colors)
plt.title('Stacked Area Plot for Temporal Data')
plt.xlabel('Date')
plt.xticks(rotation=45, ha='right') 
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

date_rng = pd.date_range(start='2022-01-01', end='2022-01-10', freq='D')
values = np.random.randint(1, 50, size=len(date_rng))

df = pd.DataFrame({'Date': date_rng, 'Values': values})

plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Values'], marker='o', linestyle='-', color='b', label='Values')
plt.title('Line Chart for Temporal Data')
plt.xlabel('Date')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
import plotly.graph_objects as go

# Sample data for two candles
open_prices = [100, 110]
close_prices = [120, 105]
high_prices = [130, 115]
low_prices = [95, 100]

# Create candlestick trace
trace = go.Candlestick(x=[1, 2],
                       open=open_prices,
                       high=high_prices,
                       low=low_prices,
                       close=close_prices)

# Create layout
layout = go.Layout(title='Simple Candlestick Plot',
                   xaxis=dict(title='Candle'),
                   yaxis=dict(title='Price'))

# Create figure
fig = go.Figure(data=[trace], layout=layout)

# Show the plot
fig.show()
```

```python
import plotly.figure_factory as ff
import pandas as pd

tasks = ["Task 1", "Task 2"]
start_dates = ['2024-03-06', '2024-03-07']
end_dates = ['2024-03-10', '2024-03-12']

fig = ff.create_gantt(df=pd.DataFrame({'Task': tasks, 'Start': start_dates, 'Finish': end_dates}))
fig.update_layout(title='Simple Gantt Chart', xaxis=dict(title='Timeline'))

fig.show()
```

```python
import plotly.express as px

# Sample data
data = {'Country': ['India', 'USA', 'UK', 'Russia', 'Egypt', 'Brazil', 'Australia'],
        'Value': [10, 15, 7, 12, 8, 5, 20]}

# Create a Geochart
fig = px.choropleth(data, locations='Country', locationmode='country names', color='Value',
                    color_continuous_scale='Viridis', title='Simple Geochart')

# Show the plot
fig.show()
```
